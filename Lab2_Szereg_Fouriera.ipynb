```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import quad
```

```python
T=2
A=-1
B=A+T
w=2*np.pi/T #=PI
NaN = float('NaN')
print(w)
def s(t):
  if(t>A and t<=B):
    return 1.36*t**2
  else:
    return NaN # funkcja jest okreÅ›lona tylko w przedziale (-1,1>
```

```python
def fa1(t):
  return s(t)*np.cos(w*t)
def fa2(t):
  return s(t)*np.cos(2*w*t)

a0=1/T*quad(s,A,B)[0]
a1=2/T*quad(fa1,A,B)[0]
a2=2/T*quad(fa2,A,B)[0]

print('a0='+str(np.round(a0,4))+' a1='+str(np.round(a1,4))+' a2='+str(np.round(a2,4)))
```

```python
def f(t):
  return a0+a1*np.cos(w*t)+a2*np.cos(2*w*t)
```

```python
a=-3
b=3
t = np.linspace(a,b, 100)
fig, ax = plt.subplots(figsize=(10,10))
# ax.plot(t,s(t),t,f(t))
y1=[s(x) for x in t]
y2=[f(x) for x in t]

ax.plot(t,y1,t,y2)
plt.legend(('s(t)','f(t)'),loc="upper right",prop={'size': 16})
ax.grid(True)
plt.show()
```

```python
def okresowa(t):
  return s((t-A) % (T) +A)

y3=[okresowa(x) for x in t]

fig, ax = plt.subplots(figsize=(10,10))
ax.plot(t,y3,t,y2)
plt.legend(('s(t)','f(t)'),loc="upper right",prop={'size': 16})
ax.grid(True)
plt.show()
```
